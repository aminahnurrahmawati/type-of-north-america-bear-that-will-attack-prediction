# let's do classification algorithm to our dataset

I will try to predict what type of bear that will attack from the data set that we have. 
first, I use Naive Bayes Algorithm, and let's see how accurate the prediction

```python
#there are still categorical data, to apply Machine Learning algorithm  to the data set we need to convert them to be numerical

data_set['Type of bear']=label_encoder.fit_transform(data_set['Type of bear'])
data_set['Location']=label_encoder.fit_transform(data_set['Location'])
data_set['Type']=label_encoder.fit_transform(data_set['Type'])
data_set['Description']=label_encoder.fit_transform(data_set['Description'])
```

```python
data = data_set.iloc[:, [1,2,6,7,8,9]]

data #these are parameters that we will use to predict the type of bear that will attack
```

```python
X = data_set.iloc[:, [1,2,6,7,8]].values
y = data_set.iloc[:, [9]].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0) #devide 80% from data set as data train and the rest is data test
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
#Training Naive Bayes on Training Data
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

```python
prediction = classifier.predict(X_test)
```

```python
#let's evaluate our model using confussion matrrix

from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test,prediction)
print(cm)
accuracy_score(y_test, prediction)
```

# this second try, I will use decision tree algorithm

```python
 # I will try to apply decision Tree

from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
model.fit(X_train, y_train)
```

```python
prediction_2= model.predict(X_test)
```

```python
#let's evaluate our model using confussion matrrix

from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test,prediction_2)
print(cm)
accuracy_score(y_test, prediction_2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier2 = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
classifier2.fit(X_train, y_train)
```

```python
prediction_knn = classifier2.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test,prediction_knn)
print(cm)
accuracy_score(y_test, prediction_knn)
```
